### <h3><a id="1a" style="color:#5575a9; text-decoration: None">a. Binomial distribution</a></h3>
<hr style="height: 2px; border: none; background-color: #DFDFEA;">

$f(k; n, p) = {n \choose k}p^k q^{n-k}$

````python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(1,2,figsize = (14,6), tight_layout=True)
axs[0].set_title('Probability distribution',fontsize = 20)
axs[0].set_xlabel('$k$',fontsize = 16)
axs[0].set_ylabel('$f(k;n,p)$',fontsize = 16)
axs[1].set_title('Cumulative distribution',fontsize = 20)
axs[1].set_xlabel('$k$',fontsize = 16)
axs[1].set_ylabel('$F(k;n,p)$',fontsize = 16)

k = np.arange(0,40)
n = 50
p = 0.4

axs[0].scatter(k, binom.pmf(k,n,p))
axs[1].scatter(k, binom.cdf(k,n,p))

mean = n*p
var = n*p*(1-p)
print(mean, var)
````

### <h3><a id="1b" style="color:#5575a9; text-decoration: None">b. Poisson distribution</a></h3>
<hr style="height: 2px; border: none; background-color: #DFDFEA;">

````python
from scipy.stats import poisson

fig, axs = plt.subplots(1,2,figsize = (14,6), tight_layout=True)
axs[0].set_title('Probability distribution',fontsize = 20)
axs[0].set_xlabel('$k$',fontsize = 16)
axs[0].set_ylabel('$f(k;\lambda)$',fontsize = 16)
axs[1].set_title('Cumulative distribution',fontsize = 20)
axs[1].set_xlabel('$k$',fontsize = 16)
axs[1].set_ylabel('$F(k;\lambda)$',fontsize = 16)

k = np.arange(0,20)
lamb = 4.1

axs[0].scatter(k, poisson.pmf(k,lamb))
axs[1].scatter(k, poisson.cdf(k,lamb))

mean = lamb
var = lamb
````

## <h2><a id="AcceptReject">Accept-Reject Methods</a></h2>

````python
def f(x):
    return -x**2 + 4 #Change this to your function - don't worry about normalisation

# We want to generate values within the range X=-2..2. 
# We need $a$ the maximum value of $f(X)$ in that range which we set by hand
a_value = f(0.0) #Change this

# For reproducibility, I like to fix the seed for the random number generator, any number is as good as any other
rng = np.random.default_rng(0) 

# Generate uniform X samples in the region of interest
min_x = -2 #Change this
max_x = 2 #Change this
x_samples = rng.uniform(min_x, max_x, size = 5000)

# Generate uniform y samples in U(0, max(f(X)) (where the max(f(X) here is at f(0))
y_samples = rng.uniform(0, a_value,size = 5000)
 
# Find which samples fulfill the criteria y <= f(X);
# This is an array of True and False values 
accept = (y_samples <= f(x_samples))

fig, ax = plt.subplots(1,1,figsize = (9,7), tight_layout=True)
ax.plot(np.linspace(min_x, max_x, 100),f(np.linspace(min_x, max_x, 100)),label='$f(X)$',color='black')
# Next plot ALL the points tried as failed points 
ax.scatter(x_samples, y_samples, s = 1, label='Rejected samples', color='#56B4E9')
# but then for the points we accept we plot another point to indicate accepted points as this will be on top and seen by us
ax.scatter(x_samples[accept],y_samples[accept],s = 1,c='#D55E00',label = 'Accepted samples')
ax.legend(fontsize = 12)
````

<hr style="border:2px solid #9EC6E1">

<div style="background-color:#DFDFEA">

**Type-I error rate:** $\alpha = \int_{\text{critical region}}f(X) \, dX$

**Power of a hypothesis test:** $1-\beta = \int_{\text{critical region}}g(X) \, dX$

**p-value**

$p = P(t|H_0)=\int_t^\infty dt' \, p(t'|H_0)$
- $p<\alpha$ Statistically significant - reject H_0
- $p>\alpha$ Not significant - fail to reject

- `ppf(alpha)` returns the value of X that gives the cumulative probability `alpha` (returns the inverse of the CDF)
- `.interval(1-alpha)` retusn the boundaries of that interval that contains $(1-\alpha)$ of the distribution: = `ppf([(1-alpha)/2, (1+alpha)/2])`
    - works for any continuous (and many discrete) distributions in scipy.stats, not just norm. SciPy defines .interval(confidence) as the equal-tailed interval. 


````python
import numpy as np
import scipy

alpha = 0.05
interval = norm.interval(1 - alpha, loc = 0, scale = 1) # Standard normal distribution
print(interval)
````

## <h2><a id="ChiSquared">Chi-squared distribution</a></h2>

Assume that the datapoints $y_{i}$ is normally dsitributed from the model function as the mean value. A good fit is where $y_{i}$ differs from the mean model value in the order of $\sigma_{i}$. Therefore, for a good fit, the chi-squared value should be close to the number of datapoints (reduced-chi-squared close to 1).

$\chi^2(y_i) = \sum_i p_i^2 = \sum_i \frac{(y_i - f(X_i))^2}{(\sigma_i)^2},
$

Choose $\alpha$ (e.g. 5%), determine if we would reject $H_0$ using p-value:

````python
from scipy.stats import chi2
chi2_value = 7
N_dof = 5

print("p-value:", chi2.sf(chi2_value, N_dof)) #func.sf()="Survival function"=(1 - func.cdf)
````

## <h2><a id="KStest">Kolmogorov-Smirnov tests</a></h2>

<div style="background-color:#DFDFEA">

**Empirical cumulative distribution** $F_{\text{data}}(X) = \frac{\text{number of data values} < X}{n},$

We can look at performing either a one-sided or a two-sided hypothesis test on the Kolmogorov-Smirnov statistic. These correlate to different null and alternate hypotheses:

* Two-sided test: Null hypothesis is $F(X) = G(X)$; alternate hypothesis is $F(X) \neq G(X)$
<br>

* One-sided test: there are two possible tests:
    1. $H_0: F(X) \geq G(X)\, \forall \, X$; $H_1: F(X) \lt G(X)$ for at least one value of $X$
    1. $H_0: F(X) \leq G(X)$ \, \forall \, $X$; $H_1: F(X) \gt G(X)$ for at least one value of $X$

- One-sided test: `scipy.stats.ksone`
- Two-sided test: `scipy.stats.kstwo`

````python
np.random.seed(0) 
number_data = 50
samples = norm.rvs(loc=0.0, scale=1.0, size = number_data)
Xmin=-3.0
Xmax=3.0
Xvalues = np.linspace(Xmin, Xmax, number_data*2) 

def Fd(X):
    """ This gives the empirical cumulative distribution of the data in "samples" for given X value """
    return np.count_nonzero(samples<X)/len(samples)

empirical = [Fd(X) for X in Xvalues] # For visualization
````

<a style="color:#5575a9;">**One sample KS test**</a>

How likely is it that we would see these data samples if the data is drawn from that probability distribution? $D = \max_X |F_\text{data}(X) - G(X)|$. In general, larger D indicate worse agreement

````python
# Calculate manually
from scipy.stats import kstwo
samples_cumulative = [Fd(X) for X in samples] # For data evaluation
D = np.max(np.abs(samples_cumulative - norm.cdf(samples)))
p_value = kstwo.sf(D,len(samples))
print(D, p_value)

# ks_1samp
from scipy.stats import ks_1samp
d1_D, d1_p = ks_1samp(samples, norm.cdf)   # Only need to input the sample.
print(d1_D, d1_p)

#kstest
from scipy.stats import kstest
ks_result = kstest(samples, norm.cdf)
print("Now D = {:.3f}, p-value = {}".format(ks_result.statistic,ks_result.pvalue))
````

<a style="color:#5575a9;">**Two sample KS test:**</a>

How likely is it that we would see these two sets of data samples if they are drawn from the same (but unknown) probability distribution?
$D_2 = \text{sup}_X|F_\text{data}(X) - G_\text{data}(X)|$

````python
# ks_2samp
from scipy.stats import ks_2samp
samples_normal_1 = norm.rvs(size = number_data)
samples_normal_2 = norm.rvs(size = number_data)
d2_D, d2_p = ks_2samp(samples_normal_1, samples_normal_2)  
print(d2_D, d2_p)

#kstest
ks_result = kstest(samples_normal_1, samples_normal_2)
print("Now D = {:.3f}, p-value = {}".format(ks_result.statistic,ks_result.pvalue))
````

# <h1><a id="Week03" style="color:black;">Week 3: Likelihoods</a></h1>

<hr style="border:2px solid #9EC6E1">

Likelihood: $P(X|H_{0})$:

$$P(H_{0}|X) = P(X|H_{0})\dfrac{P(H_{0})}{P(X)}$$

## <h2><a id="loglikelihoods">Log likelihoods</a></h2>

<div style="background-color:#DFDFEA">

$$
\ln L(H) = \ln \prod_{i}^{N}f(X_{i}|H) = \sum_{i=1}^{N} \ln f(X_{i}|H)
$$

The *likelihood* for either discrete or continuous distribution is just the product of observing the outcome.

````python
from scipy.stats import expon #exponential distribution

measurements = np.array([0.72863217, 0.30653282, 0.99725302, 1.95956204, 1.60472692, 0.48397563, 0.14558313, 4.60142088, 0.55139557, 1.30257874])

l_H = np.prod(expon.pdf(measurements))
print("Likelihood exponential distribution: {}".format(l_H))

log_l_H = np.sum(expon.logpdf(measurements))
print("log-Likelihood exponential distribution: {}".format(log_l_H))
````

## <h2><a id="binnedlikelihoods">Binned likelihoods</a></h2>

<div style="background-color:#DFDFEA">

A binned likelihood is simply a **product  of Poisson likelihoods** since the count in each bin is distributed as a Poisson random variable $\left( f(k, \lambda)=e^{-\lambda}\lambda^k / k! \right)$. If you have some probability  density function $f$, then the Poisson parameter $\lambda$ for the bin defined as the range $[x_{a},x_{b}]$ (bin-boundaries) is 
defined as:

$$
    \lambda = N\int_{x_{a}}^{x_{b}} f(x)dx
$$

````python
from scipy.stats import poisson

def PoissonLambda(a,b):
    # function to return the integral of the NORMAL distribution between a and b
    return len(observations)*(norm.sf(a) - norm.sf(b))

def Count(a,b,observations):
    # function to return the number of events between the range a and b
    return np.size(observations[(observations >= a) & (observations < b)])

bins = np.linspace(-5,5,num=10,endpoint=True)
intervals = [[bins[i],bins[i+1]] for i in range(len(bins)-1)]

# get the counts in data and integrals for each interval
counts = [Count(x[0],x[1],observations) for x in intervals]
lambdas = [PoissonLambda(x[0],x[1]) for x in intervals]

binned_ln_l = np.sum([poisson.logpmf(c,l) for c,l in zip(counts,lambdas)])   # <--- How likely you are to see bins with this count and lambda given by Poisson.

print("binned log-Likelihood normal distribution: {}".format(binned_ln_l))
````

## <h2><a id="LHR">Negative Log-Likelihood Ratio (LHR)</a></h2>

<div style="background-color:#DFDFEA">
The LHR is often calculated analytically (Neyman-Pearson lemma):

$$
-\ln(\Lambda) = -\ln \left(\dfrac{L(H_{1})}{L(H_0)}\right)  = \ln L(H_{0}) - \ln L(H_{1})
$$

</div>

<div style="background-color:#e3d8e6">

***Side note*** (optional, but may be useful):
$$
    1-\beta  = \int_{c^{\prime}}f(X|H_{1})dX

             = \int_{c^{\prime}}\Lambda f(X|H_{0})dX

             =  E\left[\Lambda \right]_{c^{\prime}},
$$
We want to choose $c'$ that maximizes $\Lambda = \frac{f(X|H_{1})}{f(X|H_{0})}$

````python
import matplotlib.pyplot as plt
from scipy.stats import norm
import pandas as pd

fig, ax = plt.subplots(1,1,figsize = (6,5),dpi = 150)
df = pd.read_csv('data_1D.csv')

n_samples = 10000
nlogLambdaDistH0 = []
nlogLambdaDistH1 = []

for i in range(n_samples):
    rvsH0 = norm.rvs(0,1,size=len(df['data'])) # Generate data under H0
    rvsH1 = norm.rvs(1,1,size=len(df['data'])) # Generate data under H1

    nlogLambdaDistH0.append(sum(0.5-rvsH0)) # Value of -ln(Lambda) for this sample under H0
    nlogLambdaDistH1.append(sum(0.5-rvsH1)) # Value of -ln(Lambda) for this sample under H1

nlogLambda = sum(0.5-(df['data'])) # Value of -ln(Lambda) for the observed data

ax.hist(nlogLambdaDistH0, bins=30, color='blue', histtype='step', density=True,label="$H_{0}$")
ax.hist(nlogLambdaDistH1, bins=30, color='orange', histtype='step', density=True,label="$H_{1}$")

ax.axvline(nlogLambda,color='black',linestyle='--',label='Observed $-\\ln(\\Lambda)$')

ax.set_xlabel('$-\\ln(\\Lambda)$',fontsize = 12)
ax.set_ylabel('Probability Density',fontsize = 12)

ax.legend()
````

# <h1><a id="Week04" style="color:black;">Week 4: Parameter Estimation and Fitting Data</a></h1>

<hr style="border:2px solid #9EC6E1">

Parameter estimations include **maximizing log-likelihood** $\color{#5575a9} \ln L(\theta; X_i) = \sum_{i = 1}^N\ln f(X_i;\theta)$ (which is found analytically as a function of the parameter) and **minimizing chi-squared** $\color{#5575a9} \left(\chi^2 = 2C - 2\ln(L) \right)$.

**Numerical optimisation of cost functions**

Then you will use methods like gradient descent to find the optimal parameter value. Recall that ***gradient descent*** is used for minimisation problems (chi-squared and negative log likelihood). This can be implemented using iMinuit.

<div style="background-color:#E3D8E6">

1. Define an initial parameter guess $\theta_{\text{init}}$ and set our parameter estimate $\hat{\theta}$ equal to it
1. Calculate (numerically) the gradient of the cost function $q$ evaluated at the initial guess, $\frac{\partial q}{\partial \hat{\theta}}$
1. Update the value of $\hat{\theta}$ according to
\begin{equation*}
\hat{\theta} = \hat{\theta} - \alpha \frac{\partial q}{\partial\hat{\theta}}
\end{equation*}
where $\alpha$ is a parameter known as the **learning rate**. 
1. Repeat steps 2 and 3 until the value of the cost function is below some satisfactory value or satisfies some convergence condition

<hr style="height: 2px; border: none; background-color: #DFDFEA;">

For **uncertainty $\Sigma$** in the parameter estimates, Taylor expand around the stationary point and rewrite the log-likehood or chi-squared in the form of a Gaussian log-likelihood or a chi-squared distribution. The term where the $\sigma$ is located will be taken as the error of the estimation.

This means the uncertainty is the range that changes the log-likelihood by -1/2 (from **maximum**):
$$\ln(L(\hat{\theta}\pm\Sigma)) = \ln(L(\hat{\theta})) - \frac{1}{2}$$

whereas for the chi-squared we look for the range that changes by +1 (from **minimum**):
$$\chi^2(\hat{\theta} \pm \Sigma) \approx \chi^2(\hat{\theta}) + 1$$

## <h2><a id="binned">Binned likelihood</a></h2>

Approximating the predicted count in a bin: 

$$\color{#5575a9} n(b) \approx (X_b - X_{b-1}) \, f(\bar{X}_{b}; \theta) \, , \quad \text{where} \quad \bar{X}_b = (X_{b} + X_{b-1})/2 \, .$$

The cost functions for binned fits that are implemented in Minuit assumes that the bin counts are independently Poisson distributed about some unkown expected value per bin. This is correct for ordinary histograms. 

For a given PDF $f(X; \boldsymbol{\theta})$ with parameters $\boldsymbol{\theta}$, we can write the expected count $\lambda_i$ in a given bin with boundaries $a_i$ and $b_i$ as

\begin{equation*}
\lambda_i = \int_a^bf(X;\boldsymbol{\theta}) dX
\end{equation*}

## <h2><a id="perfmetrics">Performance metrics in classification</a></h2>

### **Confusion matrix**<hr style="height: 2px; border: none; background-color: #C9CAE1;">


For binary classification problem with $y_i \in \{+1, -1\}$,

* **True positive (TP)**: Samples in the +1 class that are correctly identified as the +1 class
<br>

* **False positive (FP)**: Samples in the -1 class that are incorrectly identified as the +1 class (type I error)
<br>

* **True negative (TN)**: Samples in the -1 class that are correctly identified as the -1 class
<br>

* **False negative (FN)**: Samples in the +1 class that are incorrectly identified as the -1 class (type II error)

<div style="background-color:#DFDFEA">

* **Accuracy**: Proportion of samples that are classified correctly
<br>

\begin{equation*}
\text{Accuracy} = \frac{\text{Number of correct predictions}}{\text{Total number of predictions}} = \frac{TP + TN}{TP + TN + FP + FN}
\end{equation*}
<br>

* **Fall-out or False Positive Rate**: Proportion of points with $y$ = -1 that are incorrectly predicted as $\hat{y}$ = +1
<br>

\begin{equation*}
\text{Fall-out} = \frac{\text{+1 predictions with true class -1}}{\text{Points with true class -1}} = \frac{FP}{FP + TN}
\end{equation*}
<br>

* **Precision**: Proportion of points with predicted to have class +1 that have true class +1 i.e. fraction of predictions of class +1 that are correct
<br>

\begin{equation*}
\text{Precision} = \frac{\text{+1 class predictions with true class +1}}{\text{Predictions of class +1}} = \frac{TP}{TP + FP}
\end{equation*}
<br>

* **Recall or True Positive Rate**: Fraction of points with true class +1 that are correctly identified as class +1
<br>

\begin{equation*}
\text{Recall} = \frac{\text{+1 class predictions with true class +1}}{\text{Points with true class +1}} = \frac{TP}{TP + FN}
\end{equation*}

### <h3><a style="color:#5575a9; text-decoration: None">The Receiver Operating Characteristic (ROC) curve</a></h3>

The receiver operating characteristic (ROC) curve is the plot of fallout (FP rate) against recall (TP rate), as a function of decision boundary value. In general, it is used to evaluate the effectiveness of a classifier.

**AUC (area-under-curve) score**
* Complete metric for the area under the ROC curve is referred to as **ROC-AUC**
* A perfect classifier has a ROC-AUC score of 1 (ROC curve = straight line y=x)
* A random classifier has a ROC-AUC score of 0.5 
* The worst possible classifier has a ROC-AUC score of 0.

### Pre-processing the data

Sometimes, we need to perform some pre-processing on our data in order to simplify our data or make it easier for our algorithm to solve the problem:

* **Rescaling:** transforming your data into a specific range, e.g. such that all of your input features are between 0 and 1 or between -1 and 1. If the scale between different features is significantly different then one feature can influence the training more than others.
<br>

* **Encoding categorical features:** some datasets will have non-numeric data, e.g. whether a house has a garden or not, which must be transformed into numerical data for the purposes of training a machine learning algorithm.

<div style="background-color:#DFDFEA">


The cut positions are chosen to optimise some function of the model prediction on training data. For classification, we need some function that quantifies how good our data split is, i.e. how well it separates our different classes. This is given by the **Gini impurity**:

\begin{equation*}
G_i = 1 - \sum_k p_{i, k}^2,
\end{equation*}

where $\sum_k p_{i, k}^2$ is therefore the total probability of correctly labelling a randomly selected sample with a randomly selected label at node $i$, so the Gini impurity $G_i$ is the probability of incorrectly labelling a randomly selected sample.

Alternative option: **log loss** or the **entropy**:

\begin{equation*}
H_i = -\sum_k p_{i, k}\log_2\left(p_{i,k}\right)
\end{equation*}

The Gini and entropy of given node is 0 if it only has samples of one class. Regardless of which loss function we use, the model decides on the the cut at each node by minimising the loss function for that node.

### sklearn Decision Tree syntax

For a `DecisionTreeClassifier`, the `score` method finds the mean accuracy on test data, while for a `DecisionTreeRegressor` the `score` method finds the coefficient of determination ($R^2$ score).

Hyperparameters:
* `max_depth`: the maximum depth of the tree, i.e. how many layers of nodes are allowed. If the value passed is `None`, the tree continues "growing" until all leaves are pure (either only one class, or perfect prediction on training data) or `min_samples_split` is reached.

* `min_samples_split` (**evaluated before splitting**): the minimum number of samples in a set required to consider splitting an internal node, i.e. we require $n_m$ > `min_samples_split` to split the dataset $Q_m$ into two datasets.

* `min_samples_leaf` (**evaluated after splitting**): the number of samples from the training dataset required to be at a leaf node. A split at *any* depth will only be considered if it leaves at least `min_samples_leaf` training samples in each of the left and right branches.

If any of the conditions are not met, it automatically becomes a leaf.

## <h2><a style="color:#7F62AA;" id="ensemble">Ensemble Learning (code non-examinable)</a></h2>

Use *many* individual models (which may be the same type of model or different ones, e.g. a mixture of SVMs and DTs), which make their own independent predictions, and we use some **voting scheme** to determine the ensemble prediction based on the individual predictions of each model. We use different voting schemes for classification and regression, and we can do things in a couple different ways: 

* For classification, we can use either: 
    * Hard voting: the prediction is the most common prediction across all the models
    * Soft voting: the prediction is the most probable prediction across all the models i.e. predicts the class with the greatest total probability across all models in the ensemble
<br>
<br>

* For regression, we can predict either the mean prediction across all models in the ensemble, or a weighted mean. This lets us assign different weights to different models.

````python
from sklearn.ensemble import VotingClassifier
from sklearn.svm import SVC
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784',as_frame = False)
X, y = mnist['data'], mnist['target']
X_train, X_test, y_train, y_test = X[:20000],X[20000:],y[:20000],y[20000:]

svc_clf = SVC(max_iter = 20)
tree_clf = tree.DecisionTreeClassifier(max_depth = 20, random_state = 1)

voting_clf = VotingClassifier(estimators = [('svc',svc_clf),('tree',tree_clf)], voting = 'hard')

# train each classifier and evaluate their accuracy on the test set
for clf in (svc_clf, tree_clf, voting_clf):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print('{} accuracy: {}'.format(clf.__class__.__name__, accuracy_score(y_test, y_pred)))
````

<hr style="border:2px solid #9EC6E1">

**Neurons:** take sets of inputs and produce an output according to the weights and bias of that neuron: $y = \mathbf{w}\cdot\mathbf{X} + b$

**Activation functions:** $y = \sigma(\mathbf{w}\cdot\mathbf{X} + b)$. Without activation functions, we can only ever approximate linear functions or classify linearly separable data

* Sigmoid: 
$\sigma(z) = \frac{1}{1 + e^{-z}}$

* tanh: $\sigma(z) = \frac{e^z - e^{-z}}{e^z + e^{-z}}$

* ReLU (Rectified Linear Unit): $\sigma (z) = \begin{cases}
  0  & z \lt 0 \\
  z & \text{otherwise}
\end{cases}$

**Multilayer NN:** Arrange neurons into layers, the ouputs from one set of neurons = inputs to the next set. Each neuron in a layer can take inputs from *every* neuron in the previous layer with independent weight for each of these: a given layer of neurons has a *matrix* of weights, corresponding to every possible pair of input neuron and output neuron.
* Input layer (first layer) $\to$ hidden layers $\to$ output layer (last layer)
* $\color{#7F62AA} f_i(\mathbf{X}_i) = \sigma_i (\mathbf{w}_i \cdot \mathbf{X}_i + b_i) \Rightarrow \mathbf{y} = f_N\left(f_{N - 1} \left(\cdots \left(f_{1} \left(\mathbf{X}\right) \right) \right) \right)$

## <h2><a id="NNL">Neural Network Learning</a></h2>

### <h3><a id="understandnnl" style="color:#5575a9; text-decoration: None">Understanding NNL</a></h3>
<!-- <hr style="height: 2px; border: none; background-color: #DFDFEA;"> -->

**Data**
* **Training dataset:** used to train the model.

* **Validation data:** used during training to monitor generalisation of the network. Want to ensure loss on validation data improves as well as the training data; if validation loss stops improving while training loss continues to improve, this is a sign of *overfitting*.

* **Test data:** used after *all* training is finished to evaluate final performance.
<hr style="height: 2px; border: none; background-color: #DFDFEA;">

**Loss functions:** minimized wrt *all* individual weights and biases across the entire network - ***gradient descent*** wrt all parameters
* **For classification:** Require normalization (ouput = probabilities); minimize ***cross entropy*** `nn.BCELoss()`
* **For regression:** Normalization not required; minimize *mean squared error*
<hr style="height: 2px; border: none; background-color: #DFDFEA;">

**Backpropagation:** start from the last layer of the NN, calculate all of the relevant loss function derivatives, work backwards through the network
<hr style="height: 2px; border: none; background-color: #DFDFEA;">

**Stochastic gradient descent:** `torch.optim.SGD` break data into random batches, update by the gradient of the loss function averaged across the data in the batch (save time). In each epoch, we update the weights for each batch, then reshuffle the batches between epochs
<!-- <hr style="height: 2px; border: none; background-color: #DFDFEA;"> -->

<div style="background-color:#DFDFEA">
The overall procedure for training our model in a single epoch therefore goes as follows: <a id='algorithm'></a>

1. Obtain a single batch of our data
<br>

1. Perform a forward pass, i.e. calculate prediction on the training data in the batch
<br>

1. Zero the gradients on our optimizer in preparation for calculating gradients of the batch
<br>

1. Calculate the loss from the prediction and the batch training targets
<br>

1. Perform a backward pass i.e. calculate the gradients of the loss function with respect to each model parameter
<br>

1. Use the gradients to perform a step of the optimizer and update the model parameters
<br>

1. Repeat the process for the next batch until all batches have been used to update the model parameters

We will see this implemented in code in [Section Four](#code-algorithm).

<hr style="height: 2px; border: none; background-color: #E3D8E6;">

**Activation functions**
* **Vanishing gradient problem**: if the expectation of the activation function gradient is considerably $\ll 1$, for parameters far from the output $\frac{\partial L}{\partial p}$ will become extremely close to 0, resulting in very small steps every update. Results in slow or stalled learning. 
<br>

* **Exploding gradient problem**: if the expectation of the activation function gradient is $\gg 1$, $\frac{\partial L}{\partial p}$ will be very large for layers far from the output, so can cause wild fluctuations in the weights with every update by overshooting the minimum. This makes it difficult for the model to train effectively. 

Activation functions such as **sigmoid and tanh are explicitly bad** for ***deep*** neural networks because of these reasons. 

In general, a good choice for a deep network is a variant on the ReLU (issue: *dead neurons*), with such variants including:

* $
\text{LeakyReLU}(z) = \begin{cases}
  z  & z \geq 0 \\
  \alpha\cdot z & \text{otherwise}
\end{cases}$

* $
\text{ELU}(z) = \begin{cases}
z & z \gt 0 \\
\alpha\cdot (e^{z} - 1) & \text{otherwise}
\end{cases}$

* $
\text{Swish}(z) = \frac{z}{1 + e^{-z}}$
<hr style="height: 2px; border: none; background-color: #E3D8E6;">

### <h3><a id="codennl" style="color:#5575a9; text-decoration: None">Code for training NNL with PyTorch</a></h3>
<hr style="height: 2px; border: none; background-color: #DFDFEA;">

#### Example data

````python
from sklearn.datasets import make_classification
import numpy as np

n_samples = 300
np.random.seed(42)

X, y = make_classification(n_features = 2, 
                           n_redundant = 0, 
                           n_samples = n_samples, 
                           n_informative = 2, 
                           random_state = 6, 
                           class_sep = 1, 
                           n_clusters_per_class=1)

X_torch = torch.tensor(X).float()
y_torch = torch.tensor(y).float()
````

#### Model

````python
# Define model
torch.manual_seed(5)

model = nn.Sequential(nn.Linear(2,10),
                      nn.ReLU(),
                      nn.Linear(10,1),
                      nn.Sigmoid())

# Define dataset
dataset = TensorDataset(X_torch, y_torch)
dataset_train, dataset_validate, dataset_test = random_split(dataset, lengths = [0.6,0.2,0.2], generator = torch.Generator().manual_seed(2)) # Split dataset into separate datasets for training & testing

dloader_train = DataLoader(dataset_train, batch_size = 32, shuffle = True)
dloader_validate = DataLoader(dataset_validate, batch_size = 32, shuffle = True)

# Define loss function and optimizer
# loss_fcn = nn.BCELoss() # Default reduction is 'mean' which means the loss is averaged over the batch, can also set reduction to 'sum' to get the total loss for the batch.
loss_fcn = nn.BCELoss(reduction='sum')
optimizer = torch.optim.SGD(params = model.parameters(), 
                            lr = 0.01) # Other parameters include momentum, weight_decay, etc.
````

**Training Loop** - the number of times we run the training loop = number of *epochs*

````python
def train_epoch():
    tot_loss = 0
    valid_loss = 0
    for X_train, y_train in dloader_train:
        y_pred = model(X_train)
        optimizer.zero_grad() # Set the gradients to zero before calculating the new gradients for this batch, otherwise they will accumulate with the previous batch's gradients
        loss = loss_fcn(y_pred, y_train.reshape(-1,1))
        tot_loss += loss.detach()
        loss.backward()
        optimizer.step() # Update the model parameters using the calculated gradients
    
    for X_valid, y_valid in dloader_validate:
        y_pred_v = model(X_valid)
        vloss = loss_fcn(y_pred_v, y_valid.reshape(-1,1))
        valid_loss += vloss.detach()
    
    return tot_loss/len(dataset_train), valid_loss/len(dataset_validate) # Returns the average training and validation loss for this epoch
````

````python
X_train = torch.vstack([dataset_train[i][0] for i in range(len(dataset_train))])
y_train = torch.vstack([dataset_train[i][1] for i in range(len(dataset_train))])
X_valid = torch.vstack([dataset_validate[i][0] for i in range(len(dataset_validate))])
y_valid = torch.vstack([dataset_validate[i][1] for i in range(len(dataset_validate))])#
X_test = torch.vstack([dataset_test[i][0] for i in range(len(dataset_test))])
y_test = torch.vstack([dataset_test[i][1] for i in range(len(dataset_test))])


def plot_decision_boundary(ax, scatter_x, scatter_y):
    N = 1000
    X_grid = np.meshgrid(np.linspace(-7,7,N),np.linspace(-7,7,N))
    X_grid2 = np.array([X_grid[0].flatten(),X_grid[1].flatten()])
    preds = model(torch.tensor(X_grid2.transpose()).float()).reshape((N, N)).detach()
    ax.contourf(X_grid[0],X_grid[1], preds, cmap = plt.cm.cividis, alpha = 0.5)
    ax.scatter(scatter_x[:,0],scatter_x[:,1],c = scatter_y, cmap = plt.cm.cividis, edgecolor='black',lw = 0.5)
    ax.set_xlabel('$X_1$',fontsize = 16)
    ax.set_ylabel('$X_2$',fontsize = 16)
    ax.set_xlim(-4, 4)
    ax.set_ylim(-3.5,3.5)
````

#### Accuracy and ROC-AUC

**Accuracy:** if our output is greater than 0.5, we say it is more likely that the given point is in class 2 and so we predict class 2. In contrast, if the output is less than 0.5 then we predict class 1.

````python
# Need to decide what class to predict, and make sure the prediction tensor is in the right shape
from sklearn.metrics import accuracy_score

train_pred = torch.Tensor([0 if x < 0.5 else 1 for x in model(X_train)]).reshape(y_train.shape)
train_accuracy = accuracy_score(y_true = y_train,
                                y_pred = train_pred)

valid_pred = torch.Tensor([0 if x < 0.5 else 1 for x in model(X_valid)]).reshape(y_valid.shape)
valid_accuracy = accuracy_score(y_true = y_valid,
                                y_pred = valid_pred)

print('Training accuracy = {:.1f}%'.format(train_accuracy*100))
print('Validation accuracy = {:.1f}%'.format(valid_accuracy*100))
````

````python
from sklearn.metrics import roc_curve, roc_auc_score

roc_score = roc_auc_score(y_true = y_valid.detach().numpy(),
                          y_score = model(X_valid).detach().numpy())

fpr, tpr, thresholds = roc_curve(y_true = y_valid.detach().numpy(),
                                 y_score = model(X_valid).detach().numpy())

fig, ax = plt.subplots(1,1,figsize = (6, 4), dpi = 150)
ax.plot(fpr, tpr, color='#D55E00')
ax.set_xlabel('False positive rate',fontsize = 20)
ax.set_ylabel('True positive rate',fontsize = 20)
ax.set_title('ROC curve, validation data',fontsize = 24)
ax.tick_params(which='both',direction='in',top=True,right=True,labelsize = 16)
ax.grid(color='xkcd:dark blue',alpha = 0.2)

print('ROC score = {:.3f}'.format(roc_score))
````
